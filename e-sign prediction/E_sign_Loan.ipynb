```
# getting features ready
dataset = data.drop(['entry_id' , 'months_employed'] , axis = 1)
```

```
# one hot encoding
dataset = pd.get_dummies(dataset)
```

```
dataset.head()
```

```
# making X and Y of the data
y = dataset.e_signed
X = dataset.drop(['e_signed'] , axis = 1)
```

```
# train test split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 22)
```

```
# Standerdise the dataset
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.fit_transform(x_test)
```

```
# doing PCA to see stucture of data
from sklearn.decomposition import PCA
pca = PCA(x_train.shape[1])
pca.fit(x_train)
print(pca.explained_variance_ratio_)
```

```
pca = PCA(2)
reduced = pca.fit_transform(x_train)
```

```
# plot the values 
plt.scatter(reduced[:,0], reduced[:,1], c = y_train)
plt.show()
```

```
# building first model
from sklearn.linear_model import LogisticRegression
clf_lr = LogisticRegression(solver = 'lbfgs')
clf_lr.fit(x_train, y_train)
```

```
# making evaluation
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix
def evaluate(clf):
    y_pred = clf.predict(x_test)
    accuracy = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    cm = confusion_matrix(y_test, y_pred)
    print("The Scores for {} are: Accuracy of {}, F1 of {} and CM".format(type(clf).__name__, accuracy, f1))
    sns.heatmap(cm, annot = True , fmt='g')
```

```
evaluate(clf_lr)
```

```
from sklearn.svm import SVC
clf_svm = SVC(kernel = 'rbf')
clf_svm.fit(x_train,y_train) 
```

```
evaluate(clf_svm)
```

```
from sklearn.neighbors import KNeighborsClassifier
clf_knn = KNeighborsClassifier(100)
clf_knn.fit(x_train, y_train)
```

```
evaluate(clf_knn)
```

```
from sklearn.tree import DecisionTreeClassifier
clf_dt = DecisionTreeClassifier()
clf_dt.fit(x_train, y_train)
```

```
evaluate(clf_dt)
```

```
from sklearn.ensemble import RandomForestClassifier
clf_rf = RandomForestClassifier()
clf_rf.fit(x_train, y_train)
```

```
evaluate(clf_rf)
```

```
# Grid Search for SVM
parameters = {'kernel': ['rbf','linear'] }
```

```
from sklearn.model_selection import GridSearchCV
grid_search = GridSearchCV(estimator = clf_svm,
                          param_grid = parameters,
                          scoring = 'accuracy',
                          n_jobs = -1,
                          cv= 10)
```

```
from sklearn.ensemble import AdaBoostClassifier, BaggingClassifier
clf_ab = AdaBoostClassifier(n_estimators=100, learning_rate=0.3)
clf_ab.fit(x_train, y_train)
```

```
evaluate(clf_ab)
```

```
clf_bg = BaggingClassifier(n_estimators=300)
clf_bg.fit(x_train, y_train)
```

```
evaluate(clf_bg)
```

```
from sklearn.naive_bayes import BernoulliNB
clf_bnb = BernoulliNB()
clf_bnb.fit(x_train, y_train)
```

```
evaluate(clf_bnb)
```

```
# building deep neural network
from keras.models import Sequential
from keras.layers import Dense, Dropout
```

```
model = Sequential()
model.add(Dense(100 , input_dim = x_train.shape[1], activation = 'relu'))
model.add(Dropout(0.4))
model.add(Dense(200, activation = 'relu'))
model.add(Dropout(0.4))
model.add(Dense(300, activation = 'relu'))
model.add(Dropout(0.4))
model.add(Dense(1, activation = 'sigmoid'))
model.summary()
```

```
model.compile(loss = 'binary_crossentropy' , optimizer = 'rmsprop', metrics = ['accuracy'])
```

```
model.fit(x_train, y_train, epochs = 100, batch_size = 64 )
```

```
scores = model.evaluate(x_test, y_test)
print(scores[1])
```
